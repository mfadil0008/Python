### <font color ='coral'>2. Read list file to csv</font>

```python
import pandas as pd
list_file_csv = pd.read_csv('list_file.txt', sep = '_', header=None)
list_file_csv.columns=['ID','Name','Title']
list_file_csv.head()
```

### <font color ='coral'>3. Function clear: clear attribut file (.docx)</font>

```python
def clear(text):
    if '.doc' in text:
        return(text[:text.find('.doc')])
    elif '.pdf' in text:
        return(text[:text.find('.pdf')])
    else:
        return(text)
```

```python
list_file_csv = list_file_csv.applymap(clear)
list_file_csv.head()
```

### <font color ='coral'>4. Save data frame to csv</font>

```python
list_file_csv.to_csv('list_file.csv',index=None)
```
